```python
df[df[['Item_Weight']].isnull().any(axis=1)]
```

If we see one specific Item_Identifier that has null value in Item_weight, it is clear that Item_weight is same for a particular 
Item_Identifier. So we can fill the Item_Weight with the mean value by category Item_Identifier.

```python
df.groupby(['Item_Identifier'])[['Item_Weight']].mean().head()

#Another way of doing the same as above
#df.pivot_table(index='Item_Identifier',values='Item_Weight')
```

```python
df['Item_Weight'] = df['Item_Weight'].fillna(df.groupby(['Item_Identifier'])['Item_Weight'].transform('mean'))
```

All missing Item_Weight is filled out with mean value by category Item_Identifier.
Let's check with below command

```python
pd.isnull(df['Item_Weight']).sum(axis=0)
```

Still 4 records remains with Null Item_Weight.
Let's analyze them

```python
df[df[['Item_Weight']].isnull().any(axis=1)]
```

```python
df[df['Item_Identifier']=='FDN52']
```

All 4 records are the unique record as per the category Item_Identifier, that's why mean value can't be calculated.
So we have to find another way to fill in.
One way can be mean value of Item_Weight categorize by Item_Fat_Content,Item_Type

```python
df.groupby(['Item_Fat_Content','Item_Type'])[['Item_Weight']].mean()
```

```python
df['Item_Weight'] = df['Item_Weight'].fillna(df.groupby(['Item_Fat_Content','Item_Type'])['Item_Weight'].transform('mean'))
```

Now Analyze the Outlet_Size attribute which has a missing values

```python
df['Outlet_Size'].value_counts(dropna=False)
```

Since Outlet_Size is the attribute of Outlet, so it can be filled with the mode of Outlet_size as per Outlet_Identifier category 

```python
df[df[['Outlet_Size']].isnull().any(axis=1)]
```

```python
df['Outlet_Identifier'].value_counts()
```

```python
null_OS = df[df[['Outlet_Size']].isnull().any(axis=1)]
null_OS['Outlet_Identifier'].value_counts()
```

But looking at the above result, Outlet_Identifier category cannot be used to find the mode.
So let's turn the attention to Outlet_Type category to find the mode.

```python
df.groupby(['Outlet_Type'])['Outlet_Size'].agg(pd.Series.mode)
```

```python
df['Outlet_Size'] = df['Outlet_Size'].fillna(df.groupby(['Outlet_Type'])['Outlet_Size'].agg(pd.Series.mode)[0])
```

```python
pd.isnull(df['Outlet_Size']).sum(axis=0)
```

Treatment of missing value in Outlet_Size is done

Some item have 0 visibility yet they sold, so it does not make any sense. Let's impute 0 with mean visibility of that product

```python
print('Total items with zero visibility',len(df[df['Item_Visibility']==0]))

#replace zero with NaN
df['Item_Visibility'].replace(0,np.nan,inplace=True)

#Fill the na's with mean value by Item_Identifier category
df['Item_Visibility'] = df['Item_Visibility'].fillna(df.groupby(['Item_Identifier'])['Item_Visibility'].transform('mean'))

print('Null values in Item_Visibility',pd.isnull(df['Item_Visibility']).sum())
```

## Part 3 - DEVELOP

##### One-hot Encoding

```python
cat_vars = ['Item_Fat_Content','Item_Type','Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type',
            'New_Identifier']

df_copy = pd.concat([df_copy, pd.get_dummies(df_copy[cat_vars], drop_first=True)], axis=1)
df_copy.head()
```

```python
df_copy.drop(cat_vars, axis=1, inplace=True)
df_copy.dtypes
```

```python
del df_copy['Item_Identifier']
df_copy.head()
```

Move Item_Outlet_Sales to the last as a target feature

```python
sales = df_copy['Item_Outlet_Sales']
df_copy.drop('Item_Outlet_Sales', axis=1, inplace=True)
df_copy.insert(38, 'Item_Outlet_Sales', sales)
df_copy.head()
```

```python
X = df_copy.iloc[:,1:38].values
y = df_copy.iloc[:,38].values


X_train, X_test, y_train, y_test = train_test_split(X,y , test_size=0.20, random_state=42)

print("The length of X_train is ", X_train.shape)
print("The length of y_train is ", y_train.shape)
print("The length of X_test is ", X_test.shape)
print("The length of y_test is ", y_test.shape)
```

#### Feature Scaling

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### ----  Create models ----

```python
def train_model(model_name, model):
    
    print("Model is : ",model_name)
    model.fit(X_train,y_train)

    y_pred = model.predict(X_test)

    # Compute and print R^2 and RMSE
    print("R^2: {}".format(r2_score( y_test, y_pred)*100))

    mse = mean_squared_error(y_test,y_pred)
    print("Mean Squared Error: {}".format(mse))


    rmse = np.sqrt(mean_squared_error(y_test,y_pred))
    print("Root Mean Squared Error: {}".format(rmse))
    
    #scoring = {'r2': 'r2_score', 'RMSE':'neg_root_mean_squared_error'}
    #cv_scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
    
    # Print the 5-fold cross-validation scores
    #print(-1*cv_scores)

    #print("Average 5-Fold CV Score: {}".format(np.sqrt(-1*np.mean(cv_scores))))
```

```python
lm = LinearRegression(normalize=True)
rfc = RandomForestRegressor(n_estimators=200, bootstrap=True, min_samples_leaf=40, min_samples_split=8, max_depth=5)
gbr = GradientBoostingRegressor(learning_rate=0.01, n_estimators=300, max_depth=6, min_samples_split=8, min_samples_leaf=80)

model_list =[('Linear Regression',lm),('Random Forest',rfc),  ('Gradient Boosting',gbr)]

for name, i in model_list:
    print('---------------------')
    train_model(name, i)
    print("\n")
    
```

### ----  Test models ----

#### Hyperparameter Tuning

Using GridSearchCV

```python
forest = RandomForestRegressor(n_jobs=-1)

param_grid = [
{'n_estimators': [200, 300], 'min_samples_leaf':[40,50,60],
 'max_depth': [5,6], 'bootstrap': [True]}
]

grid_search_forest = GridSearchCV(forest, param_grid, cv=10, scoring='neg_mean_squared_error')
grid_search_forest.fit(X_train, y_train)
```

```python
cvres = grid_search_forest.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
grid_search_forest.best_estimator_
```

```python
grid_search_forest.best_params_
```

```python
rfc = RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=6,
           max_features='auto', max_leaf_nodes=None,
           min_impurity_decrease=0.0, min_impurity_split=None,
           min_samples_leaf=40, min_samples_split=2,
           min_weight_fraction_leaf=0.0, n_estimators=200, n_jobs=-1,
           oob_score=False, random_state=None, verbose=0, warm_start=False)
rfc.fit(X_train,y_train)

y_pred = rfc.predict(X_test)

# Compute and print R^2 and RMSE
print("R^2: {}".format(r2_score( y_test, y_pred)*100))

mse = mean_squared_error(y_test,y_pred)
print("Mean Squared Error: {}".format(mse))
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: {}".format(rmse))
```

Using Randomized Search

```python
np.linspace(1, 45, num = 3)
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 300, num = 3)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(4, 7, num = 4)]

# Minimum number of samples required to split a node
min_samples_split = [5, 10]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split}

pprint(random_grid)
```

```python
# Use the random grid to search for best hyperparameters

rf_random = RandomizedSearchCV(estimator = forest, param_distributions = random_grid, n_iter = 10, cv = 10, verbose=2, random_state=42, n_jobs = -1, scoring='neg_mean_squared_error')

# Fit the random search model
rf_random.fit(X_train, y_train)
```

```python
#now let's check how the RMSE changes for each parameter configuration
cvres2 = rf_random.cv_results_
for mean_score, params in zip(cvres2["mean_test_score"], cvres2["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
rf_random.best_estimator_
```

```python
rf_random.best_params_
```

```python
random_best= rf_random.best_estimator_.predict(X_test)
final_mse = mean_squared_error(y_test, random_best)
final_rmse = np.sqrt(final_mse)
print('The best model from the Randomized Search has a RMSE of %f and R-sq of %f'  %(round(final_rmse, 2), r2_score( y_test, y_pred)*100))
```
